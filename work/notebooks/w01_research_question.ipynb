```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Rows:", len(df))
print("Columns:", len(df.columns))

corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")

wc = df.groupby("trend_direction")["word_count"].median()
print("\nMedian word count by trend direction:")
print(wc)
```

The starter dataset contains 30,000 pages. The correlation between search volume and impressions is 0.001, indicating almost no linear relationship in this dataset. The median word count for pages with different trend directions is similar, suggesting that content length alone is not enough to explain page performance. These observations suggest that multiple signals should be considered when prioritizing pages for review.
