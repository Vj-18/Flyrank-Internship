## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Ranking (Scoring).

I chose a ranking/scoring task because the goal is not simply to classify pages as "good" or "bad." Instead, the objective is to rank web pages based on their likelihood of needing a content refresh. This helps content teams decide which pages should be reviewed first when resources are limited.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The starter dataset uses is_declining_label, which is derived from trend_direction == "down". This is a proxy label based on observed page performance rather than a future outcome. In a larger project, a stronger target would be future traffic decline or recovery measured over a later time window.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Precision@50 measures how many of the top 50 recommended pages are actually declining. A higher Precision@50 means the ranking is better at prioritizing pages that deserve review. In the starter notebook, the Random Forest model achieved a Precision@50 of 0.740, compared with 0.240 for the hand-written baseline rule.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule can only consider a few conditions at a time, such as page age or impressions. Machine learning can combine many observable signals, including impressions, CTR, content age, average position, engagement, and word count, to identify more complex patterns. The Week 1 results showed that the Random Forest model achieved a higher Precision@50 than the hand-written rule, suggesting that combining multiple signals can produce a more effective ranking for review.
